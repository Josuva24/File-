```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

def generate_ai_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    output = model.generate(**inputs, max_new_tokens=800)
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
class ColorPaletteGenerator:

    def __init__(self):
        self.palettes = {
            "modern": {
                "primary": "#6366f1",
                "secondary": "#8b5cf6",
                "accent": "#ec4899",
                "background": "#0f172a",
                "surface": "#1e293b",
                "text_primary": "#ffffff",
                "text_secondary": "#cbd5e1",
                "border": "#334155"
            },
            "ocean": {
                "primary": "#0284c7",
                "secondary": "#0ea5e9",
                "accent": "#22d3ee",
                "background": "#082f49",
                "surface": "#0c4a6e",
                "text_primary": "#ffffff",
                "text_secondary": "#bae6fd",
                "border": "#075985"
            },
            "forest": {
                "primary": "#16a34a",
                "secondary": "#15803d",
                "accent": "#22c55e",
                "background": "#052e16",
                "surface": "#14532d",
                "text_primary": "#ffffff",
                "text_secondary": "#bbf7d0",
                "border": "#166534"
            }
        }

    def get_palette(self, style):
        return self.palettes.get(style, self.palettes["modern"])
```

```python
class HTMLPageBuilder:

    def __init__(self, palette):
        self.palette = palette

    def generate_hero(self, title):
        return f"""
        <section class="hero">
            <h1>{title}</h1>
            <p>Welcome to our professional website</p>
            <button>Get Started</button>
        </section>
        """

    def generate_about(self):
        return """
        <section class="about">
            <h2>About Us</h2>
            <p>We create modern and responsive websites.</p>
        </section>
        """

    def generate_contact(self):
        return """
        <section class="contact">
            <h2>Contact Us</h2>
            <form>
                <input type="text" placeholder="Your Name" required>
                <input type="email" placeholder="Your Email" required>
                <textarea placeholder="Message"></textarea>
                <button type="submit">Send</button>
            </form>
        </section>
        """

    def build_page(self, title):
        hero = self.generate_hero(title)
        about = self.generate_about()
        contact = self.generate_contact()

        return f"""
        <!DOCTYPE html>
        <html>
        <head>
            <title>{title}</title>
            <style>
                {generate_css(self.palette)}
            </style>
        </head>
        <body>
            {hero}
            {about}
            {contact}
        </body>
        </html>
        """
```

```python
def generate_css(colors):
    return f"""
    body {{
        margin: 0;
        font-family: Arial, sans-serif;
        background: {colors['background']};
        color: {colors['text_primary']};
    }}

    section {{
        padding: 60px 20px;
        text-align: center;
    }}

    .hero {{
        background: linear-gradient(135deg, {colors['primary']}, {colors['secondary']});
        color: white;
    }}

    button {{
        background: {colors['accent']};
        color: white;
        border: none;
        padding: 10px 20px;
        cursor: pointer;
        border-radius: 6px;
    }}

    button:hover {{
        opacity: 0.8;
    }}

    input, textarea {{
        display: block;
        margin: 10px auto;
        padding: 10px;
        width: 80%;
        border-radius: 5px;
        border: 1px solid {colors['border']};
    }}

    @media (max-width: 768px) {{
        section {{
            padding: 30px 10px;
        }}
    }}
    """
```

```python
import gradio as gr

palette_generator = ColorPaletteGenerator()

def generate_page(title, description, style):
    palette = palette_generator.get_palette(style)
    builder = HTMLPageBuilder(palette)
    html_code = builder.build_page(title)
    return html_code

with gr.Blocks() as app:
    gr.Markdown("# HTML Quick-Styler")

    with gr.Tab("Generate Page"):
        title = gr.Textbox(label="Page Title")
        description = gr.Textbox(label="Description")
        style = gr.Dropdown(["modern", "ocean", "forest"], label="Color Style")
        output = gr.Code(language="html")
        btn = gr.Button("Generate")
        btn.click(generate_page, [title, description, style], output)

    with gr.Tab("Live Preview"):
        preview = gr.HTML()
        btn.click(generate_page, [title, description, style], preview)

    with gr.Tab("Color Palettes"):
        gr.Markdown("Available Styles: modern, ocean, forest")

    with gr.Tab("Templates"):
        gr.Markdown("Hero, About, Contact templates included.")

    with gr.Tab("Guide"):
        gr.Markdown("1. Enter Title\n2. Select Style\n3. Click Generate")

app.launch(share=True)
```
